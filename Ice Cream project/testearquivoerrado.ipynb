```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('IceCreamData.csv')
df['Temperature'] = df['Temperature'].round()
df['Revenue'] = df['Revenue'].round()

df
```

```python
plt.plot(df['Temperature'], df['Revenue'], 'r.-')
plt.ylabel('Sales US$')
plt.xlabel('Temperatura(fahrenheit)')
plt.title('Ice Cream Truck')
plt.yticks(df['Revenue'][250:550:50])
plt.xticks(df['Temperature'][1:100:10])
```
