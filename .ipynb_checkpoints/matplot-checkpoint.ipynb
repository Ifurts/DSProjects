```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
gas = pd.read_csv('gas_prices.csv')
gas
```

```python
plt.title('Gas price comparsion')
#plt.plot (gas['Year'], gas ['Canada'],'r.-', label = 'Canada')
#plt.plot (gas['Year'], gas ['USA'],'b.-', label = 'USA')
#plt.plot (gas['Year'], gas ['France'],'y.-', label = 'France')
#ao inves de criar uma linha para cada pais, pode ser feito um loop com if e uma lista de paises, so adicionar o nome do pais na lista

list_country = ['Canada', 'USA', 'France', 'Germany']

for country in gas:
    if country in list_country:
         plt.plot(gas['Year'], gas[country], '.-', label=[country])

plt.ylabel('US$')
plt.xlabel('Year')
plt.xticks(gas['Year'][::2])
plt.legend()
plt.savefig('country.png')
plt.show()

```
